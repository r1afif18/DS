```python
# ============================================
# STEP-1 — CLEANING + FEATURE MATRIX (ANTI-LEAK) — Colab Ready, Anti-Error
# ============================================

!pip -q install gdown openpyxl xlrd

import os, io, re, json, warnings, requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

# ---------------- CONFIG ----------------
GAIKINDO_URL = "https://s.id/Dataset-LombaPeDaS2025"
DATE_START   = "2021-01-01"
DATE_END     = "2025-06-01"
HIST_CUTOFF  = "2024-12-01"   # hanya sampai sini kita boleh pakai untuk imputasi/statistik
FREQ         = "MS"
OUT_DIR      = "/content/step1_outputs"
RAW_DIR      = "/content"     # tempat kamu upload file eksternal (opsional)
PROC_DIR     = "/content/external/processed"

os.makedirs(OUT_DIR, exist_ok=True)
os.makedirs(PROC_DIR, exist_ok=True)

def log(msg):
    print(msg, flush=True)

# ---------- Downloader (s.id → Google Drive) ----------
def download_or_resolve(url: str, target="/content/dataset_from_url.bin", timeout=120):
    log("== STEP 1: Load data")
    log("Downloading from shortlink...")
    r = requests.get(url, allow_redirects=True, timeout=timeout)
    r.raise_for_status()
    ctype = (r.headers.get("Content-Type") or "").lower()
    data  = r.content

    if ("text/html" not in ctype) and (b"<html" not in data[:2048].lower()):
        with open(target, "wb") as f: f.write(data)
        return target, ctype

    body = r.text
    candidates = []
    if "drive.google.com" in r.url:
        candidates.append(r.url)
    for m in re.findall(r'https?://drive\.google\.com/[^\s\'"]+', body):
        candidates.append(m)
    for m in re.findall(r'id=([a-zA-Z0-9_-]{10,})', body):
        candidates.append(f"https://drive.google.com/uc?id={m}")

    if not candidates:
        raise RuntimeError("Shortlink mengarah ke HTML dan tidak ada tautan Google Drive di dalamnya.")

    import gdown
    gd_url = candidates[0]
    id_match = re.search(r"/d/([a-zA-Z0-9_-]+)", gd_url) or re.search(r"id=([a-zA-Z0-9_-]+)", gd_url)
    if id_match:
        file_id = id_match.group(1)
        gd_dl = f"https://drive.google.com/uc?id={file_id}"
    else:
        gd_dl = gd_url

    log("Resolved to Google Drive → downloading via gdown…")
    gdown.download(gd_dl, target, quiet=False)
    return target, "application/octet-stream"

# ---------- Reader (CSV/Excel autodetect yang beneran robust) ----------
def try_read_any(path):
    # Coba CSV dengan beberapa separator
    for sep in [None, ",", ";", "\t", "|"]:
        try:
            df = pd.read_csv(path, sep=sep, engine="python" if sep is None else "c")
            if df.shape[1] > 1:
                return df
        except Exception:
            pass
    # Coba Excel
    for engine in ["openpyxl", "xlrd"]:
        try:
            df = pd.read_excel(path, engine=engine)
            if df.shape[1] >= 1:
                return df
        except Exception:
            pass
    raise RuntimeError("Gagal baca file sebagai CSV/Excel (mungkin HTML/Zip).")

# ---------- Date detector (tahan angka YYYYMM juga) ----------
def detect_date_col(df: pd.DataFrame):
    lower = {c: str(c).lower() for c in df.columns}
    keys = ["date","tanggal","bulan","month","periode","period","waktu","time","tahun","yr","year"]
    # exact match
    for k in keys:
        for c in df.columns:
            if lower[c] == k: return c
    # partial match
    for k in keys:
        for c in df.columns:
            if k in lower[c]: return c
    # test parsing
    for c in df.columns:
        ser = df[c].dropna().head(10)
        try:
            pd.to_datetime(ser, errors="raise"); return c
        except Exception:
            pass
    # fallback
    nonnum = [c for c in df.columns if not pd.api.types.is_numeric_dtype(df[c])]
    return nonnum[0] if nonnum else df.columns[0]

# ---------- Parser tanggal tahan format ----------
def parse_dates_robust(s: pd.Series):
    """
    Mencoba: %Y-%m, %Y/%m, %Y.%m, %Y %m, %Y-%m-%d,
             integer/string YYYYMM, YYYYMMDD, dll.
    """
    x = s.copy()

    # Jika numeric 6/8 digit (YYYYMM / YYYYMMDD)
    if pd.api.types.is_integer_dtype(x) or pd.api.types.is_float_dtype(x):
        xx = pd.to_numeric(x, errors="coerce").dropna().astype(int).astype(str)
        if (xx.str.len()==6).all():
            # YYYYMM → set tgl 01
            return pd.to_datetime(xx + "01", format="%Y%m%d", errors="coerce")
        if (xx.str.len()==8).all():
            return pd.to_datetime(xx, format="%Y%m%d", errors="coerce")

    # Jika string, coba beberapa pola
    xs = x.astype(str).str.strip()
    # Normalisasi separator
    xs_norm = xs.str.replace(r"[./]", "-", regex=True).str.replace(r"\s+", "-", regex=True)

    # Coba YYYY-MM atau YYYY-MM-DD
    dt = pd.to_datetime(xs_norm, errors="coerce", format=None)
    if dt.notna().sum() >= len(xs)*0.7:
        return dt

    # Coba pakai infer lagi
    dt2 = pd.to_datetime(xs, errors="coerce", infer_datetime_format=True, dayfirst=True)
    return dt2

# ---------- Build monthly frame ----------
def ensure_monthly_frame(df, date_col, start=DATE_START, end=DATE_END, freq=FREQ):
    tmp = df.copy()
    parsed = parse_dates_robust(tmp[date_col])
    tmp = tmp.assign(**{date_col: parsed})
    tmp = tmp.dropna(subset=[date_col]).drop_duplicates(subset=[date_col])
    tmp = tmp.set_index(date_col).sort_index()
    # Pastikan index datetime
    if not isinstance(tmp.index, pd.DatetimeIndex):
        tmp.index = pd.to_datetime(tmp.index, errors="coerce")
        tmp = tmp.dropna(axis=0, subset=[tmp.index.name])
        tmp = tmp.sort_index()
    full_idx = pd.date_range(pd.Timestamp(start), pd.Timestamp(end), freq=freq)
    tmp = tmp.reindex(full_idx)
    tmp.index.name = "date"
    return tmp

# ---------- Imputasi historis saja (anti-leak) ----------
def fill_missing_historical_only(df, brand_cols, hist_end=HIST_CUTOFF):
    out = df.copy()
    # pastikan index datetime & sorted
    out.index = pd.to_datetime(out.index, errors="coerce")
    out = out.sort_index()

    flags = pd.DataFrame(False, index=out.index,
                         columns=[f"{b}_was_imputed" for b in brand_cols])
    hist_end_ts = pd.Timestamp(hist_end)
    hist_mask = out.index <= hist_end_ts

    for b in brand_cols:
        s_hist = out.loc[hist_mask, b]
        was_na_hist = s_hist.isna()
        if was_na_hist.any():
            s_filled = (s_hist
                        .interpolate("linear", limit_direction="both")
                        .ffill()
                        .bfill())
            out.loc[hist_mask, b] = s_filled
            flags.loc[was_na_hist.index, f"{b}_was_imputed"] = was_na_hist.values
        out[b] = out[b].clip(lower=0)
    return out, flags

# ---------- IQR capping (statistik historis saja) ----------
def iqr_bounds(s: pd.Series, mult=2.5):
    x = s.dropna()
    if len(x) < 4: return None, None
    q1, q3 = x.quantile([0.25, 0.75]); iqr = q3 - q1
    return q1 - mult*iqr, q3 + mult*iqr

def cap_outliers_historical(df, brand_cols, hist_end=HIST_CUTOFF, mult=2.5):
    out = df.copy()
    out.index = pd.to_datetime(out.index, errors="coerce")
    out = out.sort_index()

    flags = pd.DataFrame(False, index=out.index, columns=[f"{b}_is_outlier" for b in brand_cols])
    hist_mask = out.index <= pd.Timestamp(hist_end)
    hist = out.loc[hist_mask, brand_cols]

    for b in brand_cols:
        lo, hi = iqr_bounds(hist[b], mult)
        if lo is None:
            continue
        mask_hi = out[b] > hi
        mask_lo = out[b] < lo
        flags.loc[mask_hi | mask_lo, f"{b}_is_outlier"] = True
        out.loc[mask_hi, b] = hi
        out.loc[mask_lo, b] = lo
        out[b] = out[b].clip(lower=0)
    return out, flags

# ---------- Kalender, lags, shares ----------
def ramadan_eid_calendar(index: pd.DatetimeIndex):
    ramadan = {2021:[4,5], 2022:[4,5], 2023:[3,4], 2024:[3,4], 2025:[3]}
    eid     = {2021:[5],   2022:[5],   2023:[4],   2024:[4],   2025:[3]}
    idx = pd.to_datetime(index)
    cal = pd.DataFrame(index=idx)
    cal["month"] = idx.month
    cal["quarter"] = idx.quarter
    cal["year_end_push"] = idx.month.isin([11,12]).astype(int)
    cal["is_ramadan_month"] = [1 if (d.year in ramadan and d.month in ramadan[d.year]) else 0 for d in idx]
    cal["is_eid_month"]     = [1 if (d.year in eid     and d.month in eid[d.year])     else 0 for d in idx]
    m_dum = pd.get_dummies(cal["month"], prefix="m", drop_first=True)
    q_dum = pd.get_dummies(cal["quarter"], prefix="q", drop_first=True)
    return pd.concat([cal, m_dum, q_dum], axis=1)

def make_lag_roll(df, brand_cols):
    out = pd.DataFrame(index=df.index)
    for b in brand_cols:
        out[f"{b}_lag1"]  = df[b].shift(1)
        out[f"{b}_lag3"]  = df[b].shift(3)
        out[f"{b}_lag6"]  = df[b].shift(6)
        out[f"{b}_lag12"] = df[b].shift(12)
        out[f"{b}_roll3"]  = df[b].rolling(3,  min_periods=1).mean()
        out[f"{b}_roll6"]  = df[b].rolling(6,  min_periods=1).mean()
        out[f"{b}_roll12"] = df[b].rolling(12, min_periods=1).mean()
    return out

def compute_shares(df, brand_cols):
    total = df[brand_cols].sum(axis=1)
    eps = 1e-9
    shares_df = pd.DataFrame({b: (df[b] / (total + eps))*100.0 for b in brand_cols}, index=df.index)
    share_t    = shares_df.add_suffix("_share_t_pct")
    share_lag1 = shares_df.shift(1).add_suffix("_share_t_lag1_pct")  # gunakan ini untuk modeling
    return share_t, share_lag1

# ---------- Quality checks ----------
def quality_checks(df_raw, df_final, brand_cols):
    exp_idx = pd.date_range(pd.Timestamp(DATE_START), pd.Timestamp(DATE_END), freq=FREQ)
    contig  = (len(df_final.index) == len(exp_idx)) and df_final.index.equals(exp_idx)
    nonneg  = (df_final[brand_cols] >= 0).all().all()
    # hitung missing di data RAW historis dan 2025
    try:
        raw_idx = parse_dates_robust(df_raw[detect_date_col(df_raw)])
        tmp_raw = df_raw.copy()
        tmp_raw["_parsed_date_"] = raw_idx
        tmp_raw = tmp_raw.dropna(subset=["_parsed_date_"]).set_index("_parsed_date_").sort_index()
        miss_hist_pre  = tmp_raw.loc[:pd.Timestamp(HIST_CUTOFF), brand_cols].isna().sum().to_dict()
        miss_2025_orig = tmp_raw.loc[pd.Timestamp("2025-01-01"):pd.Timestamp(DATE_END), brand_cols].isna().sum().to_dict()
    except Exception:
        miss_hist_pre, miss_2025_orig = {}, {}
    print("\n== DATA QUALITY CHECKS")
    print(f" - continuous_monthly_index: {'PASS' if contig else 'FAIL'}")
    print(f" - no_negative_sales: {'PASS' if nonneg else 'FAIL'}")
    if miss_hist_pre:
        print(f"\nMissingness s.d. 2024 (PRE-imputation): {miss_hist_pre}")
        print(f"Missingness 2025 (anti-leak, dibiarkan sesuai data asli): {miss_2025_orig}")

# ---------- EXOG (opsional) ----------
def find_first(path_root, substrings):
    for root, dirs, files in os.walk(path_root):
        for f in files:
            name = f.lower()
            if any(sub.lower() in name for sub in substrings):
                return os.path.join(root, f)
    return None

def _to_ms_index(df, date_col=None):
    if date_col is None: date_col = detect_date_col(df)
    df = df.copy()
    df[date_col] = parse_dates_robust(df[date_col])
    df = df.dropna(subset=[date_col]).drop_duplicates(subset=[date_col]).set_index(date_col).sort_index()
    df = df.resample("MS").mean()
    df = df.loc[pd.Timestamp(DATE_START):pd.Timestamp(DATE_END)]
    df.index.name = "date"
    return df

def _clamp_by_hist(df, cols, hist_end=HIST_CUTOFF, p_low=0.01, p_high=0.99):
    hist = df.loc[:pd.Timestamp(hist_end), cols]
    lo = hist.quantile(p_low)
    hi = hist.quantile(p_high)
    out = df.copy()
    for c in cols:
        out[c] = out[c].clip(lower=lo[c], upper=hi[c])
    return out

def load_bi_rate():
    path = find_first(RAW_DIR, ["bi-7day", "7day rr", "bi7", "7day"])
    if path is None:
        log("[EXOG] BI-Rate not found — skipped.")
        return None
    try:
        df  = pd.read_excel(path)
        dcol = detect_date_col(df)
        num  = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
        pick = None
        for c in num:
            if re.search(r"(7|seven).*day|rr|rate|bi", str(c), flags=re.I):
                pick = c; break
        if pick is None and num: pick = num[-1]
        tmp = df[[dcol, pick]].rename(columns={dcol:"date", pick:"birate"})
        tmp = _to_ms_index(tmp, "date")
        out = pd.DataFrame(index=tmp.index)
        out["birate"]     = tmp["birate"]
        out["birate_d1"]  = out["birate"].diff(1)
        out["birate_yoy"] = out["birate"].diff(12)
        out = out.shift(1).rename(columns={"birate":"birate_t1","birate_d1":"birate_d1_t1","birate_yoy":"birate_yoy_t1"})
        return out
    except Exception as e:
        log(f"[EXOG] BI-Rate parse error: {e} — skipped.")
        return None

def load_inflasi():
    path = find_first(RAW_DIR, ["inflasi", "cpi", "ihk"])
    if path is None:
        log("[EXOG] Inflasi not found — skipped.")
        return None
    try:
        df = pd.read_excel(path)
        dcol = detect_date_col(df)
        tmp = df.copy()
        tmp[dcol] = parse_dates_robust(tmp[dcol])

        cand_yoy = [c for c in tmp.columns if re.search(r"yoy|year.?on.?year|inflasi.?yoy", str(c), flags=re.I)]
        cand_mom = [c for c in tmp.columns if re.search(r"mom|m.?o.?m|inflasi.?mom",   str(c), flags=re.I)]
        cand_cpi = [c for c in tmp.columns if re.search(r"\bcpi\b|ihk|indeks harga",   str(c), flags=re.I)]

        use_yoy = cand_yoy[0] if cand_yoy else None
        use_mom = cand_mom[0] if cand_mom else None
        if use_yoy is None and cand_cpi:
            tmp["_cpi_level_"] = pd.to_numeric(tmp[cand_cpi[0]], errors="coerce")
            tmp["_yoy_"] = tmp["_cpi_level_"].pct_change(12)*100.0; use_yoy = "_yoy_"
        if use_mom is None and cand_cpi:
            tmp["_mom_"] = tmp[cand_cpi[0]].pct_change(1)*100.0; use_mom = "_mom_"

        keep = [dcol] + [c for c in [use_yoy, use_mom] if c is not None]
        tmp  = tmp[keep].rename(columns={dcol:"date", use_yoy:"cpi_yoy", use_mom:"cpi_mom"})
        tmp  = _to_ms_index(tmp, "date")
        out  = tmp.shift(1).rename(columns={"cpi_yoy":"cpi_yoy_t1", "cpi_mom":"cpi_mom_t1"})
        return out
    except Exception as e:
        log(f"[EXOG] Inflasi parse error: {e} — skipped.")
        return None

def load_usd_idr():
    path = find_first(RAW_DIR, ["usd", "kurs", "idr", "historical"])
    if path is None:
        log("[EXOG] USD/IDR not found — skipped.")
        return None
    try:
        df   = pd.read_csv(path)
        dcol = detect_date_col(df)
        pick = None
        for c in df.columns:
            if pd.api.types.is_numeric_dtype(df[c]) and re.search(r"price|close|rate|kurs|value|harga|last", str(c), flags=re.I):
                pick = c; break
        if pick is None:
            nums = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
            if not nums:
                log("[EXOG] USD/IDR: no numeric — skipped.")
                return None
            pick = nums[0]
        tmp = df[[dcol, pick]].rename(columns={dcol:"date", pick:"usdir"})
        tmp["date"] = parse_dates_robust(tmp["date"])
        tmp = tmp.dropna(subset=["date"]).set_index("date").sort_index().resample("MS").mean()
        out = pd.DataFrame(index=tmp.index)
        out["usdir_avg"] = tmp["usdir"]
        out["usdir_d1"]  = out["usdir_avg"].diff(1)
        out["usdir_yoy"] = out["usdir_avg"].diff(12)
        out = out.loc[pd.Timestamp(DATE_START):pd.Timestamp(DATE_END)]
        out = out.shift(1).rename(columns={"usdir_avg":"usdir_avg_t1", "usdir_d1":"usdir_d1_t1", "usdir_yoy":"usdir_yoy_t1"})
        return out
    except Exception as e:
        log(f"[EXOG] USD/IDR parse error: {e} — skipped.")
        return None

def load_brent_from_fred():
    path = find_first(RAW_DIR, ["fredgraph", "brent", "oil", "petroleum"])
    if path is None:
        log("[EXOG] Brent (FRED) not found — skipped.")
        return None
    try:
        df   = pd.read_excel(path)
        dcol = detect_date_col(df)
        num  = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
        if not num:
            log("[EXOG] Brent: no numeric — skipped.")
            return None
        pick = num[0]
        tmp  = df[[dcol, pick]].rename(columns={dcol:"date", pick:"brent"})
        tmp  = _to_ms_index(tmp, "date")
        out  = pd.DataFrame(index=tmp.index)
        out["brent_log"] = np.log(tmp["brent"].replace({0: np.nan})).replace([np.inf,-np.inf], np.nan)
        out["brent_d1"]  = tmp["brent"].pct_change(1)
        out = out.shift(1).rename(columns={"brent_log":"brent_log_t1", "brent_d1":"brent_d1_t1"})
        return out
    except Exception as e:
        log(f"[EXOG] Brent parse error: {e} — skipped.")
        return None

def build_exog_master():
    parts = []
    for loader in [load_bi_rate, load_inflasi, load_usd_idr, load_brent_from_fred]:
        x = loader()
        if x is not None:
            parts.append(x)
    if not parts:
        log("[EXOG] No exogenous features created.")
        return None
    exog = pd.concat(parts, axis=1)
    cols = [c for c in exog.columns if pd.api.types.is_numeric_dtype(exog[c])]
    exog = _clamp_by_hist(exog, cols, HIST_CUTOFF)
    exog = exog.loc[pd.Timestamp(DATE_START):pd.Timestamp(DATE_END)]
    exog.index.name = "date"
    outpath = os.path.join(PROC_DIR, "exog_master.csv")
    exog.to_csv(outpath, index=True, float_format="%.6f")
    log(f"[EXOG] Built & saved: {outpath} | cols: {list(exog.columns)}")
    return exog

# ---------- VISUALISASI ----------
def plot_eda(df_sales, brand_cols, title_suffix=""):
    # Trend
    plt.figure(figsize=(12,6))
    for b in brand_cols:
        plt.plot(df_sales.index, df_sales[b], label=b, linewidth=1.8)
    plt.title(f"Trend Penjualan per Brand {title_suffix}")
    plt.xlabel("Bulan"); plt.ylabel("Unit")
    plt.grid(alpha=0.3); plt.legend(); plt.show()

    # Profil musiman per brand
    for b in brand_cols:
        mavg = df_sales[b].groupby(df_sales.index.month).mean()
        plt.figure(figsize=(9,4))
        plt.bar(range(1,13), [mavg.get(m, np.nan) for m in range(1,13)])
        plt.title(f"Profil Musiman Bulanan — {b} {title_suffix}")
        plt.xlabel("Bulan (1..12)"); plt.ylabel("Rata-rata Unit")
        plt.grid(axis="y", alpha=0.3); plt.show()

    # Korelasi antar brand
    if len(brand_cols) > 1:
        corr = df_sales[brand_cols].corr()
        plt.figure(figsize=(6,5))
        im = plt.imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)
        plt.colorbar(im, fraction=0.046, pad=0.04)
        plt.xticks(range(len(brand_cols)), brand_cols, rotation=45, ha="right")
        plt.yticks(range(len(brand_cols)), brand_cols)
        plt.title("Korelasi Antar Brand")
        for i in range(len(brand_cols)):
            for j in range(len(brand_cols)):
                plt.text(j, i, f"{corr.iloc[i,j]:.2f}", ha="center", va="center", color="black", fontsize=9)
        plt.tight_layout(); plt.show()

    # Evolusi market share
    total = df_sales[brand_cols].sum(axis=1).replace(0, np.nan)
    shares = df_sales[brand_cols].div(total, axis=0)*100.0
    plt.figure(figsize=(12,6))
    plt.stackplot(df_sales.index, [shares[b].fillna(0) for b in brand_cols], labels=brand_cols)
    plt.title(f"Evolusi Market Share (%) {title_suffix}"); plt.xlabel("Bulan"); plt.ylabel("Share (%)")
    plt.legend(loc="upper left"); plt.show()

# ======================
# MAIN EXECUTION
# ======================
# 1) Download & read
path, ctype = download_or_resolve(GAIKINDO_URL)
df_raw = try_read_any(path)

print(f"Sumber data : {path}")
print(f"Content-Type: {ctype}")
print(f"Shape awal  : {df_raw.shape}")
print(f"Kolom awal  : {list(df_raw.columns)}")

# 2) Deteksi tanggal & brand
date_col = detect_date_col(df_raw)
print(f"Kolom tanggal terdeteksi: {date_col}")

# numeric brand cols (exclude date column & obvious 'total')
all_numeric = [c for c in df_raw.columns if c != date_col and pd.api.types.is_numeric_dtype(df_raw[c])]
brand_cols = [c for c in all_numeric if not re.search(r"\btotal\b|\bgrand\b|\ball\b|\bmarket\b", str(c), flags=re.I)]
if not brand_cols:
    brand_cols = all_numeric[:]  # fallback safe
print(f"Brand terdeteksi: {brand_cols}")

# 3) Monthly frame, imputasi historis, outlier cap
df = ensure_monthly_frame(df_raw, date_col, DATE_START, DATE_END, FREQ)
df_imp, impute_flags   = fill_missing_historical_only(df, brand_cols, HIST_CUTOFF)
df_cap, outlier_flags  = cap_outliers_historical(df_imp, brand_cols, HIST_CUTOFF, mult=2.5)

# 4) Fitur
cal        = ramadan_eid_calendar(df_cap.index)
lags_rolls = make_lag_roll(df_cap, brand_cols)
share_t, share_lag1 = compute_shares(df_cap, brand_cols)

# 5) Exogenous (opsional; otomatis skip jika tidak ada file)
exog = build_exog_master()

# 6) Master matrix (anti-leak)
parts = [df_cap[brand_cols], impute_flags, outlier_flags, cal, lags_rolls, share_lag1]
if exog is not None:
    parts.append(exog)
master = pd.concat(parts, axis=1)

# 7) Quality checks
quality_checks(df_raw, master, brand_cols)

# 8) Save artefak
df_cap.to_csv(os.path.join(OUT_DIR, "step1_sales_capped.csv"), index=True, float_format="%.2f")
impute_flags.to_csv(os.path.join(OUT_DIR, "step1_flags_imputation.csv"), index=True)
outlier_flags.to_csv(os.path.join(OUT_DIR, "step1_flags_outlier.csv"), index=True)
cal.to_csv(os.path.join(OUT_DIR, "step1_calendar_features.csv"), index=True)
lags_rolls.to_csv(os.path.join(OUT_DIR, "step1_lag_rolling.csv"), index=True, float_format="%.2f")
share_t.to_csv(os.path.join(OUT_DIR, "step1_share_t_pct.csv"), index=True, float_format="%.2f")
share_lag1.to_csv(os.path.join(OUT_DIR, "step1_share_t_lag1_pct.csv"), index=True, float_format="%.2f")
master.to_csv(os.path.join(OUT_DIR, "step1_master_features.csv"), index=True, float_format="%.6f")

print("\n✅ STEP 1 selesai.")
print(f"📁 CSV tersimpan di: {OUT_DIR}")
print("▶️ Modeling: gunakan *_share_t_lag1_pct + lag/rolling (anti-leak).")
print("▶️ Validasi: Jan–Jun 2025 (biarkan sesuai data asli).")

# 9) Visualisasi INLINE
print("\n== PLOTS (INLINE) ==")
plot_eda(df_cap[brand_cols], brand_cols, title_suffix="(2021–2025)")

print("\n🔍 MASTER tail(10):")
try:
    from IPython.display import display
    display(master.tail(10).round(2))
except:
    print(master.tail(10).round(2))
```

```python
# ============================================
# STEP 2 — BASELINE FORECASTING (Colab Ready)
# ============================================

import os, re, numpy as np, pandas as pd
import matplotlib.pyplot as plt

STEP1_DIR = "/content/step1_outputs"
SERIES_PATH = os.path.join(STEP1_DIR, "step1_sales_capped.csv")  # hasil Step 1
DATE_START = "2021-01-01"
DATE_END   = "2025-06-01"
VAL_START  = "2025-01-01"
VAL_END    = "2025-06-01"

# -----------------------------
# UTILITIES
# -----------------------------
def smape(y_true, y_pred, eps=1e-9):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    mask = ~np.isnan(y_true) & ~np.isnan(y_pred)
    if mask.sum() == 0:
        return np.nan
    yt = y_true[mask]
    yp = y_pred[mask]
    denom = (np.abs(yt) + np.abs(yp)).clip(min=eps)
    return 100.0 * np.mean(np.abs(yp - yt) / denom)

def read_series(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Series file not found: {path}")
    df = pd.read_csv(path, index_col=0, parse_dates=True)
    # pastikan MS
    df.index = pd.DatetimeIndex(df.index)
    df = df.asfreq("MS")
    return df

def detect_brand_cols(df):
    # numeric only, exclude obvious non-brand names
    candidates = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
    brand_cols = [c for c in candidates if not re.search(r"total|share|lag|roll|imputed|outlier|ramadan|eid|m_|q_|year_end", str(c), flags=re.I)]
    # fallback jika hasil kosong
    if not brand_cols:
        brand_cols = candidates
    return brand_cols

def make_validation_index(full_index, start=VAL_START, end=VAL_END):
    idx = pd.date_range(pd.Timestamp(start), pd.Timestamp(end), freq="MS")
    # pastikan hanya bulan yang ada di series
    idx = idx.intersection(full_index)
    return idx

def build_baseline_predictions(series, brand_cols, val_index):
    """
    Kembalikan dict of DataFrames (preds per model untuk index validasi)
    dan juga prediksi JUL/AGT 2025 untuk tiap model (multi-step).
    """
    preds = {m: pd.DataFrame(index=val_index, columns=brand_cols, dtype=float)
             for m in ["naive", "snaive", "ma3", "ma6"]}

    # ----- VALIDATION PREDICTIONS (Jan–Jun 2025) -----
    for b in brand_cols:
        s = series[b]

        # naive: yhat_t = y_{t-1}
        preds["naive"][b]  = s.shift(1).reindex(val_index)

        # seasonal naive: yhat_t = y_{t-12}
        preds["snaive"][b] = s.shift(12).reindex(val_index)

        # MA3 & MA6: rolling mean sampai t-1 (anti-leak)
        preds["ma3"][b] = s.rolling(window=3, min_periods=1).mean().shift(1).reindex(val_index)
        preds["ma6"][b] = s.rolling(window=6, min_periods=1).mean().shift(1).reindex(val_index)

    # ----- MULTI-STEP FORECAST Jul & Aug 2025 -----
    fidx_jul = pd.DatetimeIndex([pd.Timestamp("2025-07-01")])
    fidx_aug = pd.DatetimeIndex([pd.Timestamp("2025-08-01")])

    # Container untuk forecast Jul/Ags per model
    f_jul = {m: pd.Series(index=brand_cols, dtype=float) for m in preds.keys()}
    f_aug = {m: pd.Series(index=brand_cols, dtype=float) for m in preds.keys()}

    for b in brand_cols:
        s = series[b].copy()
        # pastikan ada nilai sampai Jun 2025
        if pd.Timestamp("2025-06-01") not in s.index:
            continue

        # --- Jul 2025 ---
        # naive_jul = actual_jun
        naive_jul = s.loc["2025-06-01"]

        # snaive_jul = actual(Jul-12) => Jul 2024
        snaive_jul = s.loc["2024-07-01"] if "2024-07-01" in s.index.astype(str) else np.nan

        # ma3_jul = mean(May, Jun, Apr) (t-1 rolling)
        ma3_jul = s.loc[: "2025-06-01"].iloc[-3:].mean()  # mean of Apr–Jun 2025
        # ma6_jul = mean(last 6 months up to Jun)
        ma6_jul = s.loc[: "2025-06-01"].iloc[-6:].mean()

        f_jul["naive"][b]  = max(0.0, float(naive_jul))
        f_jul["snaive"][b] = max(0.0, float(snaive_jul)) if not pd.isna(snaive_jul) else np.nan
        f_jul["ma3"][b]    = max(0.0, float(ma3_jul))
        f_jul["ma6"][b]    = max(0.0, float(ma6_jul))

        # --- Aug 2025 (multi-step) ---
        # naive_aug = y_{Jul} (pred)
        naive_aug = f_jul["naive"][b]

        # snaive_aug = actual(Aug-12) => Aug 2024
        snaive_aug = s.loc["2024-08-01"] if "2024-08-01" in s.index.astype(str) else np.nan

        # ma3_aug = mean(Jul_pred, Jun_actual, May_actual)
        may = s.loc["2025-05-01"]
        jun = s.loc["2025-06-01"]
        ma3_aug = np.nanmean([f_jul["ma3"][b], jun, may])

        # ma6_aug = mean(Feb–Jun actual + Jul_pred)
        # Ambil 5 bulan actual terakhir: Feb–Jun 2025
        last5 = s.loc["2025-02-01":"2025-06-01"].values
        if len(last5) < 5:
            # fallback: ambil sebanyak mungkin terakhir
            last5 = s.loc[: "2025-06-01"].iloc[-5:].values
        ma6_aug = np.nanmean(np.append(last5, f_jul["ma6"][b]))

        f_aug["naive"][b]  = max(0.0, float(naive_aug))  if not pd.isna(naive_aug)  else np.nan
        f_aug["snaive"][b] = max(0.0, float(snaive_aug)) if not pd.isna(snaive_aug) else np.nan
        f_aug["ma3"][b]    = max(0.0, float(ma3_aug))    if not pd.isna(ma3_aug)    else np.nan
        f_aug["ma6"][b]    = max(0.0, float(ma6_aug))    if not pd.isna(ma6_aug)    else np.nan

    # Konversi ke DataFrame berindex tanggal untuk konsistensi
    fc_jul = {m: pd.DataFrame(f_jul[m]).T for m in f_jul}
    for m in fc_jul: fc_jul[m].index = fidx_jul

    fc_aug = {m: pd.DataFrame(f_aug[m]).T for m in f_aug}
    for m in fc_aug: fc_aug[m].index = fidx_aug

    return preds, fc_jul, fc_aug

def evaluate_smape(series, preds, val_index, brand_cols):
    # SMAPE per brand & model
    rows = []
    for b in brand_cols:
        y = series.loc[val_index, b].values
        for m, dfm in preds.items():
            yhat = dfm.loc[val_index, b].values
            rows.append({"brand": b, "model": m, "SMAPE": smape(y, yhat)})
    smape_df = pd.DataFrame(rows)

    # Pivot (brand x model)
    smape_pivot = smape_df.pivot(index="brand", columns="model", values="SMAPE")
    # Best per brand
    best_model = smape_pivot.idxmin(axis=1)
    best_table = smape_pivot.copy()
    best_table["best"] = best_model

    # Overall: equal & volume-weighted
    eq_overall = pd.DataFrame(
        smape_df.groupby("model")["SMAPE"].mean().sort_values().rename("SMAPE")
    ).reset_index()

    # volume-weighted pakai total penjualan aktual di val period
    vol = series.loc[val_index, brand_cols].sum()
    vol_w = smape_df.merge(
        vol.rename("weight").reset_index().rename(columns={"index":"brand"}),
        on="brand", how="left"
    )
    # normalisasi weight per brand -> rata across models sama
    vol_w["weight"] = vol_w["weight"] / vol_w.groupby("brand")["weight"].transform("first")
    weighted = (vol_w["SMAPE"] * vol_w["weight"])
    vw = pd.DataFrame(
        (weighted.groupby(vol_w["model"]).sum() / vol_w.groupby("model")["weight"].sum()).rename("SMAPE")
    ).sort_values("SMAPE").reset_index()

    return smape_pivot, eq_overall, vw, best_model

def bar_smape_per_brand(smape_pivot):
    for b in smape_pivot.index:
        plt.figure(figsize=(6,3.6))
        vals = smape_pivot.loc[b].dropna().sort_values()
        plt.bar(vals.index, vals.values)
        plt.title(f"SMAPE Validasi Jan–Jun 2025 — {b}")
        plt.ylabel("SMAPE (%)")
        plt.grid(axis="y", alpha=0.3)
        plt.tight_layout()
        plt.show()

def bar_overall(eq_overall, vw_overall):
    fig, ax = plt.subplots(1,2, figsize=(10,4))
    ax[0].bar(eq_overall["model"], eq_overall["SMAPE"])
    ax[0].set_title("Overall SMAPE — Equal-weighted")
    ax[0].set_ylabel("SMAPE (%)"); ax[0].grid(axis="y", alpha=0.3)

    ax[1].bar(vw_overall["model"], vw_overall["SMAPE"])
    ax[1].set_title("Overall SMAPE — Volume-weighted")
    ax[1].grid(axis="y", alpha=0.3)

    plt.tight_layout()
    plt.show()

def plot_val_fit(series, preds, brand_cols, val_index):
    for b in brand_cols:
        plt.figure(figsize=(10,4))
        plt.plot(series.index, series[b], label="Actual", linewidth=1.8, alpha=0.7)
        for m, dfm in preds.items():
            plt.plot(val_index, dfm[b].loc[val_index], marker="o", linestyle="--", label=m)
        plt.title(f"Actual vs Prediksi (Validation Jan–Jun 2025) — {b}")
        plt.xlabel("Bulan"); plt.ylabel("Unit")
        plt.xlim(series.index.min(), series.index.max())
        plt.grid(alpha=0.3); plt.legend(ncol=4, fontsize=8)
        plt.tight_layout()
        plt.show()

def finalize_august(series, fc_aug, best_model, brand_cols):
    """
    Ambil pemenang baseline per brand (berdasarkan SMAPE validasi) untuk prediksi Agustus 2025.
    """
    rows = []
    for b in brand_cols:
        bm = best_model.get(b, None)
        if bm is None:
            rows.append({"brand": b, "best_model": None, "forecast_2025_08": np.nan})
            continue
        f = fc_aug[bm][b].iloc[0] if b in fc_aug[bm].columns else np.nan
        rows.append({"brand": b, "best_model": bm, "forecast_2025_08": f})
    return pd.DataFrame(rows)

def bar_forecast_aug(final_df):
    ok = final_df.dropna(subset=["forecast_2025_08"])
    if ok.empty:
        print("[WARN] Tidak ada forecast valid untuk Agustus 2025.")
        return
    plt.figure(figsize=(7,4))
    plt.bar(ok["brand"], ok["forecast_2025_08"])
    for i, v in enumerate(ok["forecast_2025_08"]):
        plt.text(i, v, f"{v:,.0f}", ha="center", va="bottom", fontsize=9)
    plt.title("Prediksi Agustus 2025 — Pemenang Baseline per Brand")
    plt.ylabel("Unit")
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()

# -----------------------------
# LOAD & PREP
# -----------------------------
series = read_series(SERIES_PATH)
brand_cols = detect_brand_cols(series)
series = series[brand_cols]  # keep sales only
series = series.loc[pd.Timestamp(DATE_START):pd.Timestamp(DATE_END)]

print(f"Brands: {brand_cols}")
print(f"Date range: {series.index.min().date()} → {series.index.max().date()}")

val_index = make_validation_index(series.index, VAL_START, VAL_END)
print(f"Validation months: {len(val_index)} | {val_index.min().date()} → {val_index.max().date()}")

# -----------------------------
# FIT BASELINES & EVALUATE
# -----------------------------
preds, fc_jul, fc_aug = build_baseline_predictions(series, brand_cols, val_index)
smape_by_brand, eq_overall, vw_overall, best_model = evaluate_smape(series, preds, val_index, brand_cols)

print("\n===== SMAPE validasi Jan–Jun 2025 (per brand) =====")
display(smape_by_brand.round(2))

print("\n===== Overall SMAPE (equal-weighted) =====")
display(eq_overall.round(2))

print("\n===== Overall SMAPE (volume-weighted) =====")
display(vw_overall.round(2))

# -----------------------------
# VISUALISASI (INLINE)
# -----------------------------
bar_smape_per_brand(smape_by_brand)
bar_overall(eq_overall, vw_overall)
plot_val_fit(series, preds, brand_cols, val_index)

# -----------------------------
# FINAL — FORECAST AUG 2025
# -----------------------------
final_df = finalize_august(series, fc_aug, best_model, brand_cols)
final_df_sorted = final_df.sort_values("brand")
print("\n===== FINAL FORECAST — AUG 2025 (Baseline Winner per Brand) =====")
display(final_df_sorted.assign(forecast_2025_08=lambda d: d["forecast_2025_08"].round(2)))

# Simpan hasil step 2
OUT_DIR = "/content/step2_outputs"
os.makedirs(OUT_DIR, exist_ok=True)
smape_by_brand.to_csv(os.path.join(OUT_DIR, "step2_smape_by_brand.csv"))
eq_overall.to_csv(os.path.join(OUT_DIR, "step2_smape_overall_equal.csv"), index=False)
vw_overall.to_csv(os.path.join(OUT_DIR, "step2_smape_overall_weighted.csv"), index=False)
final_df_sorted.to_csv(os.path.join(OUT_DIR, "step2_forecast_aug2025_baseline_winner.csv"), index=False)

bar_forecast_aug(final_df_sorted)
print(f"\n📁 Outputs ditulis ke: {OUT_DIR}")
```

```python
# -----------------------------
# STEP 3: Advanced (ETS, SARIMA, LGBM)
# -----------------------------
def ets_fit_forecast(series, train_end, steps):
    y = series.copy()
    train_mask = (y.index <= pd.Timestamp(train_end))
    y_train = y.loc[train_mask].dropna()
    if len(y_train) < 18: return None, None
    try:
        mdl = ExponentialSmoothing(
            y_train, trend="add", seasonal="add", seasonal_periods=12, initialization_method="estimated"
        ).fit(optimized=True)
        start_fc = pd.Timestamp(train_end) + pd.offsets.MonthBegin(1)
        fc = mdl.forecast(steps)
        fc.index = pd.date_range(start_fc, periods=steps, freq=FREQ)
        fc = fc.clip(lower=0.0)
        return mdl, fc
    except Exception:
        return None, None

def sarima_fit_forecast(series, train_end, steps):
    y = series.clip(lower=0.0)
    train_mask = (y.index <= pd.Timestamp(train_end))
    y_train = y.loc[train_mask].dropna()
    if len(y_train) < 24: return None, None
    best_aic, best_mod = np.inf, None
    best_order, best_seasonal = None, None
    for p in [0,1]:
        for d in [0,1]:
            for q in [0,1]:
                for P in [0,1]:
                    for D in [0,1]:
                        for Q in [0,1]:
                            try:
                                mod = SARIMAX(y_train, order=(p,d,q), seasonal_order=(P,D,Q,12),
                                              trend="n", enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)
                                if mod.aic < best_aic:
                                    best_aic = mod.aic; best_mod = mod
                                    best_order, best_seasonal = (p,d,q), (P,D,Q,12)
                            except:
                                pass
    if best_mod is None: return None, None
    try:
        start_fc = pd.Timestamp(train_end) + pd.offsets.MonthBegin(1)
        fc = best_mod.get_forecast(steps=steps).predicted_mean
        fc.index = pd.date_range(start_fc, periods=steps, freq=FREQ)
        fc = fc.clip(lower=0.0)
        return best_mod, fc
    except:
        return None, None

def try_load_exog(dir_path, index_ext):
    if not (USE_EXOG and os.path.isdir(dir_path)):
        return pd.DataFrame(index=index_ext), pd.DataFrame(index=index_ext)
    frames = []
    for fn in os.listdir(dir_path):
        if not (fn.lower().endswith(".csv") or fn.lower().endswith(".xlsx")):
            continue
        path = os.path.join(dir_path, fn)
        try:
            if fn.lower().endswith(".csv"):
                tmp = pd.read_csv(path)
            else:
                tmp = pd.read_excel(path, engine="openpyxl")
        except:
            continue
        # detect date col
        cand = [c for c in tmp.columns if any(k in c.lower() for k in ["date","time","month","periode","bulan","tahun"])]
        dcol = cand[0] if cand else tmp.columns[0]
        try:
            tmp[dcol] = pd.to_datetime(tmp[dcol], errors="coerce", dayfirst=True)
        except:
            tmp[dcol] = pd.to_datetime(tmp[dcol], errors="coerce")
        tmp = tmp[~tmp[dcol].isna()].copy().set_index(dcol).sort_index()
        num = tmp.select_dtypes(include=[np.number])
        if num.shape[1]==0:
            continue
        num_m = num.resample("MS").mean()
        base = re.sub(r"[^A-Za-z0-9]+","_", os.path.splitext(fn)[0]).strip("_")
        num_m = num_m.add_prefix(base+"_")
        num_m = num_m.reindex(index_ext).ffill()
        frames.append(num_m)
    if not frames:
        return pd.DataFrame(index=index_ext), pd.DataFrame(index=index_ext)
    exog = pd.concat(frames, axis=1)
    exog_lag1 = exog.shift(1)
    return exog, exog_lag1

def build_static_features(index_ext, sales):
    cal = build_calendar(index_ext)
    total = sales.sum(axis=1).replace(0, np.nan)
    shares = (sales.div(total, axis=0))*100.0
    shares = shares.add_suffix("_share_t_pct").shift(1).add_suffix("_lag1")
    shares = shares.reindex(index_ext)
    return pd.concat([cal, shares], axis=1)

def lgbm_fit_predict(series, brand, features_static_ext, exog_lag1_ext, train_end, val_index):
    """
    - Fit ≤ 2024-12; valid Jan–Jun 2025 (hanya baris target teramati)
    - Fitur: calendar + shares_lag1 + exog_lag1 + target lag/rolling (dibentuk dinamis)
    - Recursive: pred JUL (inject) → pred AUG
    """
    y_all = series[brand].copy()
    X_full = pd.concat([features_static_ext, exog_lag1_ext], axis=1)
    X_full = add_target_lagroll(X_full, series, brand)

    # keep kolom yg ada sinyal di train
    mask_train_ext = (X_full.index <= pd.Timestamp(train_end))
    cols_keep = [c for c in X_full.columns if X_full.loc[mask_train_ext, c].notna().sum() >= 6]
    if not cols_keep:
        return None, None, None, None
    X_full = X_full[cols_keep]

    # align target ke index fitur (FIX)
    y_align = y_all.reindex(X_full.index)

    # index
    train_idx = X_full.index[(X_full.index <= pd.Timestamp(train_end)) & (~y_align.isna())]
    val_idx_int = [t for t in val_index if (t in X_full.index) and (not pd.isna(y_align.loc[t]))]
    if len(train_idx)<24 or X_full.shape[1]==0 or len(val_idx_int)==0:
        return None, None, None, None

    X_train = X_full.loc[train_idx]; y_train = y_align.loc[train_idx]
    X_valid = X_full.loc[val_idx_int]

    model = LGBMRegressor(
        n_estimators=1200, learning_rate=0.03, num_leaves=31,
        subsample=0.9, colsample_bytree=0.9, reg_lambda=1.0, random_state=42
    )
    model.fit(X_train, y_train)
    yhat_valid = pd.Series(model.predict(X_valid), index=X_valid.index)

    # recursive JUL & AUG
    tmp = series[brand].copy()
    jul, aug = pd.Timestamp("2025-07-01"), pd.Timestamp("2025-08-01")

    def assemble_X_for(t, s_current):
        feat = pd.concat([features_static_ext, exog_lag1_ext], axis=1)
        feat = add_target_lagroll(feat, s_current.to_frame(brand), brand)
        # pastikan kolom sama
        for c in X_train.columns:
            if c not in feat.columns:
                feat[c] = 0.0
        feat = feat[X_train.columns]
        return feat.loc[t:t]

    X_jul = assemble_X_for(jul, tmp)
    if X_jul.empty:
        return model, yhat_valid, None, None
    yhat_jul = float(model.predict(X_jul)[0]); tmp.loc[jul] = max(0.0, yhat_jul)

    X_aug = assemble_X_for(aug, tmp)
    if X_aug.empty:
        return model, yhat_valid, yhat_jul, None
    yhat_aug = float(model.predict(X_aug)[0])

    return model, yhat_valid, yhat_jul, yhat_aug

# EXEC Step 3
print("\n== STEP 3: Advanced")
idx_ext = pd.date_range(pd.Timestamp(DATE_START), pd.Timestamp(EXT_END), freq=FREQ)
feature_static_ext = build_static_features(idx_ext, df_cap[brands])
exog, exog_lag1 = try_load_exog(EXOG_DIR, idx_ext)

val_index = pd.date_range(pd.Timestamp(VAL_START), pd.Timestamp(VAL_END), freq=FREQ)
steps_needed = len(val_index)+2  # valid 6 + (Jul, Aug)

pred_valid = {m: pd.DataFrame(index=val_index, columns=brands, dtype=float) for m in ["ETS","SARIMA","LGBM"]}
fc_aug     = {m: pd.Series(index=brands, dtype=float) for m in ["ETS","SARIMA","LGBM"]}

for b in brands:
    y = df_cap[b].copy()

    # ETS
    ets_model, ets_fc = ets_fit_forecast(y, TRAIN_END, steps_needed)
    if ets_fc is not None:
        pred_valid["ETS"][b] = ets_fc.iloc[:len(val_index)].values
        fc_aug["ETS"][b] = float(ets_fc.iloc[len(val_index)+1])

    # SARIMA
    sar_model, sar_fc = sarima_fit_forecast(y, TRAIN_END, steps_needed)
    if sar_fc is not None:
        pred_valid["SARIMA"][b] = sar_fc.iloc[:len(val_index)].values
        fc_aug["SARIMA"][b] = float(sar_fc.iloc[len(val_index)+1])

    # LGBM
    mdl, yhat_val, yhat_jul, yhat_aug = lgbm_fit_predict(df_cap[brands], b, feature_static_ext, exog_lag1, TRAIN_END, val_index)
    if yhat_val is not None:
        pred_valid["LGBM"][b] = yhat_val
    if yhat_aug is not None:
        fc_aug["LGBM"][b] = max(0.0, float(yhat_aug))

# SMAPE advanced
rows=[]
for b in brands:
    ytrue = df_cap.loc[val_index, b].values
    for m in ["ETS","SARIMA","LGBM"]:
        rows.append({"brand": b, "model": m, "SMAPE": smape(ytrue, pred_valid[m][b].values)})
smape_adv = pd.DataFrame(rows)
print("\n===== SMAPE per Brand (Advanced, Jan–Jun 2025) =====")
print(smape_adv.pivot(index="brand", columns="model", values="SMAPE").round(2).to_string())

overall_eq_adv = smape_adv.groupby("model")["SMAPE"].mean().sort_values().reset_index(name="SMAPE")
print("\n===== Overall SMAPE (Advanced, equal-weighted) =====")
print(overall_eq_adv.round(2).to_string(index=False))

# Final Aug per brand (advanced winner + fallback)
ranked_adv = smape_adv.sort_values(["brand","SMAPE"]).groupby("brand")["model"].apply(list)
adv_rows=[]
for b in brands:
    val = np.nan
    used = "NA"
    for m in ranked_adv.get(b, []):
        v = fc_aug[m].get(b, np.nan)
        if np.isfinite(v):
            val = float(v); used = m; break
    if not np.isfinite(val):
        # fallback = nilai Juni (naive last)
        val = float(df_cap.loc[pd.Timestamp("2025-06-01"), b])
        used = "NAIVE_LAST"
    adv_rows.append({"brand": b, "best_model_adv": used, "forecast_2025_08": val})
final_adv = pd.DataFrame(adv_rows).sort_values("brand")
final_adv.to_csv(os.path.join(STEP3_DIR, "step3_advanced_final_aug2025.csv"), index=False)

print("\n===== FINAL FORECAST — AUG 2025 (Advanced Winner per Brand + Fallback) =====")
print(final_adv.to_string(index=False))
```

```python

# -----------------------------
# STEP 4: Model Selection & Submission
# -----------------------------
# Gabungkan baseline vs advanced → pilih SMAPE validasi terkecil per brand
best_bl_per_brand = smape_bl.sort_values(["brand","SMAPE"]).groupby("brand").first().reset_index()
best_adv_per_brand = smape_adv.sort_values(["brand","SMAPE"]).groupby("brand").first().reset_index()

# Map model→pred val untuk akses cepat
def _pred_val_lookup(pred_dict, model, brand):
    try:
        return pred_dict[model][brand].values
    except:
        return None

final_rows=[]
for b in brands:
    # value SMAPE baseline vs advanced
    s_bl = best_bl_per_brand.loc[best_bl_per_brand["brand"]==b, "SMAPE"].values
    s_adv= best_adv_per_brand.loc[best_adv_per_brand["brand"]==b, "SMAPE"].values
    m_bl = best_bl_per_brand.loc[best_bl_per_brand["brand"]==b, "model"].values
    m_adv= best_adv_per_brand.loc[best_adv_per_brand["brand"]==b, "model"].values
    s_bl = float(s_bl[0]) if len(s_bl)>0 else np.inf
    s_adv= float(s_adv[0]) if len(s_adv)>0 else np.inf
    mb   = m_bl[0] if len(m_bl)>0 else None
    ma   = m_adv[0] if len(m_adv)>0 else None

    # ambil forecast Aug
    f_bl = final_bl.loc[final_bl["brand"]==b, "forecast_2025_08"].values
    f_bl = float(f_bl[0]) if len(f_bl)>0 and np.isfinite(f_bl[0]) else np.nan

    f_adv= final_adv.loc[final_adv["brand"]==b, "forecast_2025_08"].values
    f_adv= float(f_adv[0]) if len(f_adv)>0 and np.isfinite(f_adv[0]) else np.nan

    if s_adv < s_bl and np.isfinite(f_adv):
        final_rows.append({"brand": b, "source": "advanced", "model": ma, "SMAPE": s_adv, "forecast_2025_08": f_adv})
    elif np.isfinite(f_bl):
        final_rows.append({"brand": b, "source": "baseline", "model": mb, "SMAPE": s_bl, "forecast_2025_08": f_bl})
    else:
        # fallback terakhir
        val = float(df_cap.loc[pd.Timestamp("2025-06-01"), b])
        final_rows.append({"brand": b, "source": "fallback", "model": "NAIVE_LAST", "SMAPE": np.nan, "forecast_2025_08": val})

final_choice = pd.DataFrame(final_rows).sort_values("brand")
print("\n===== FINAL CHOICE — per Brand (Bandingkan SMAPE Validasi) =====")
print(final_choice.to_string(index=False))

# Submission CSV sesuai format panitia: id, Prediksi_Penjualan
sub = final_choice.copy()
sub["id"] = sub["brand"].str.upper() + "-2025-08"
sub = sub[["id", "forecast_2025_08"]].rename(columns={"forecast_2025_08":"Prediksi_Penjualan"})
sub.to_csv(os.path.join(STEP4_DIR, "submission_tim.csv"), index=False)

print(f"\n📁 Submission disimpan: {os.path.join(STEP4_DIR, 'submission_tim.csv')}")

# Plot final bar
plt.figure(figsize=(8,4))
plt.bar(final_choice["brand"], final_choice["forecast_2025_08"])
for i, v in enumerate(final_choice["forecast_2025_08"]):
    plt.text(i, v, f"{v:,.0f}", ha="center", va="bottom", fontsize=9)
plt.title("Final Forecast — Agustus 2025 (per Brand)")
plt.ylabel("Unit")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout(); plt.show()

# Panel tail + titik Aug
fig, axes = plt.subplots(len(brands), 1, figsize=(10, 1.8*len(brands)), sharex=True)
if len(brands) == 1: axes = [axes]
for ax, b in zip(axes, brands):
    tail = df_cap[b].iloc[-12:]
    ax.plot(tail.index, tail.values, marker="o", linewidth=1.8, label="Actual")
    v = final_choice.loc[final_choice["brand"]==b, "forecast_2025_08"]
    if len(v)>0 and np.isfinite(v.values[0]):
        ax.plot(pd.Timestamp("2025-08-01"), float(v.values[0]), marker="^", markersize=8, linestyle="none", label="Pred Aug")
    ax.set_title(b)
    ax.grid(True, alpha=0.3)
axes[0].legend()
plt.tight_layout(); plt.show()
```
